Tensors are similar to Numpy's ndarrays. 
Uninitialized matrix is declared but does not contain definite known values before it is used.





```
import torch

#Constructing a matrix of order mxn, uninitialized.
x = torch.empty(6,5)
print(x)
```

```
#Constructing a matrix randomly initialized.
x = torch.rand(6,5)
print(x)
```

```
#Contruct a matrix filled with zero & datatype long
x = torch.zeros(6, 5, dtype=torch.long)
print(x)
```

```
#Constructing tensor directly from data
x = torch.tensor([10.5, 3.4, 4.6])
print(x)
```

```
#Create a tensor based on existing tensor. You can reuse the properties of input tensor.  
x = x.new_ones(6,5, dtype=torch.double)
print(x)
```

```
x = torch.randn_like(x, dtype=torch.float)
print(x)
```

```
#Getting size of the tensor

print(x.size())
```

Operations on Tensors

```
#Addition Operation with different syntax
x = torch.rand(3,4)
y = torch.rand(3,4)
print('\n\n--------------- X ---------------\n')
print(x)
print('\n\n--------------- Y ---------------\n')
print(y)
print('\n\n================== X+Y ==================\n')
print(x+y)
print('\n\n================== Using add() ==================\n')
print(torch.add(x,y))
print('\n\n======= Using add() with output tensor, torch.add(x, y, out=res) =========\n')
res = torch.empty(3,4)
torch.add(x, y, out=res)
print(res)
print('\nWe can also assign the output tensor as z = x+y \n')
z = x+y
print(z)
print('\n\nOR with z = torch.add(x,y) \n')
z = torch.add(x,y)
print(z)
```

```
#inplace addition using *_.
#Any operation that mutates tensor will be post-fixed with an _
print('\n\n*********** In-place addition ***********')
print('\n -------- Y(Before) ------------\n')
print(y)
y.add_(x)
print('\n\n -------- Y(After) ----------\n')
print(y)
```

Converting a Torch array to Numpy array


```
a = torch.randn(5)
print("\nTensor : ", a)

b = a.numpy()
print("\nNumpy array : ", b)

#Both a and b share their memory locations, and hence modifying one will change the other.
#(when torch tensor is on CPU) 

a.add_(1)
print('\n', a)
print('\n', b)
```

Converting a Numpy array to Torch array

```
import numpy as np

a = np.arange(5)
print('\nNumpy array : ', a)

b = torch.from_numpy(a)
print('\nTensor array : ', b)

np.add(a, 2, out=a)
print('\n', a)
print('\n', b)
```
